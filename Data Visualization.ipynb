# Data Visualization

### Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=pd.read_csv("C:\\Users\\Larry Williams\\Desktop\\Mamata\\Edvancer\\data\\data\\heart.csv")
```

### Visuvalizing the plots for one continous variable

```python
plt.hist(x["age"])
```

```python
sns.distplot(x["age"])
```

```python
plt.boxplot(x["age"])
```

```python
sns.boxplot(x["age"])
```

```python
a=pd.Series(x['age']).plot(kind="density")
```

### Visuvalizing the plots for one discrete variable

```python
plt.bar(x["cp"],height=0.5)
```

```python
sns.countplot(x["cp"])
```

```python
plt.bar(x["thal"],height=0.2)
```

```python
sns.countplot(x["thal"])
```

### Visuvalizing the plots for two continous variables

```python
plt.scatter(x["age"],x["oldpeak"])
```

```python
sns.regplot(x["age"],x["oldpeak"],fit_reg=False)
```

```python
sns.lineplot(x["age"],x["oldpeak"])
```

### Visuvalizing the plot for one continous and one discrete variables

```python
a=x.boxplot(column="age",by="thal")
```

```python
b=x.boxplot(column="oldpeak",by="cp")
```

### Visuvalizing the plot for two discrete vaiables

```python
sns.countplot(x="sex",hue="target",data=x)
```

```python
sns.countplot(x="sex",hue="target",data=x,dodge=False)
```

### Ploting multiple plots

```python
fig,axes=plt.subplots(ncols=5,figsize=(25,5))
sns.countplot(x["sex"],ax=axes[0],palette="Blues")
sns.countplot(x["cp"],ax=axes[1],palette="Set2")
sns.countplot(x["fbs"],ax=axes[2],palette="viridis")
sns.countplot(x["restecg"],ax=axes[3],palette="summer")
sns.countplot(x["exang"],ax=axes[4],palette="winter")
```

```python
fig,axes=plt.subplots(ncols=4,figsize=(25,5))
sns.countplot(x['slope'],ax=axes[0],palette="spring")
sns.countplot(x["ca"],ax=axes[1],palette="muted")
sns.countplot(x["thal"],ax=axes[2],palette="magma")
sns.countplot(x["target"],ax=axes[3],palette="copper")
```
